```python
from graphviz import Digraph
```

```python
system = Digraph()
```

```python
N = 4
```

```python
nodes = ['X'+str(n) for n in range(N)]
nodel_labels = ['a','b','c','d'] #nodes
nodes
```

```python
for n, l in zip(nodes, nodel_labels):
    system.node(n)
```
